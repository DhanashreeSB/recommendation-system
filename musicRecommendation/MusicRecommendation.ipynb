```python
def comicGraphic():
    for i,row in music.iterrows():
        if(row[' amazement']==1 and row[' power']==1 and row[' joyful_activation']==1 and row[' calmness']==1):
            print(row['track id'])
            
def travel():
    print('inside travel')
    for i,row in music.iterrows():
        if(row[' amazement']==1 and row[' power']==1 and row[' joyful_activation']==1 and row[' calmness']==1 and row[' solemnity']==1):
            print(row['track id'])
            
def fiction():
    for i,row in music.iterrows():
        if(row[' amazement']==1 & row[' power']==1 & row[' joyful_activation']==1 & row[' calmness']==1 & row[' tension']==1):
            print(row['track id']) 
            
def mystery():
    print('inside mystery')
    for i,row in music.iterrows():
        if(row[' amazement']==1 and row[' solemnity']==1):
            print(row['track id'])
            
def romance():
    print('inside romance')
    for i,row in music.iterrows():
        if(row[' tenderness']==1 and row[' nostalgia']==1 and row[' amazement']==0 and row[' power']==0 and row[' joyful_activation']==0 and row[' calmness']==0 and row[' tension']==0 and row[' solemnity']==0):
            print(row['track id'])
            
def religion():
    print('inside religion')
    for i,row in music.iterrows():
        if(row[' calmness']==1 ):
            print(row['track id'])

            
    
def recommendMusic(book_index):
    category = ""
    index=int(book_index)
    print(book_index)
    if index>=1 and index<=3026 :
        print('comic')
        category = 'Comics & Graphic Novels'
    if index>=22918 and index<=30496:
        print('literature')
        category = 'Literature & Fiction'
    if index>=186599 and index<=204936:    
        category = 'Travel'
    if index>=11730 and index<=16020 :
        print('romance')
        category = 'Romance'
    
    switcher = { 
        'Comics & Graphic Novels': comicGraphic,
        'Literature & Fiction': fiction,
        'Travel': travel,
        'Romance':romance
    }
    print(category)
    func=switcher.get(category,lambda :'Invalid')
    return func()
```
